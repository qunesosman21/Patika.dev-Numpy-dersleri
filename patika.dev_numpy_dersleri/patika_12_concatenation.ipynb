```python
import numpy as np
```

```python
a=np.array([1,2,3,4])
```

```python
b=np.array([5,6,7,8])
```

```python
# a ve b güncellenmez
np.concatenate([a,b]) # birleştirme yapar
```

```python
# python listelerinide concatanate ile birleştirebiliriz

l=[9,10]
```

```python
np.concatenate([a,b,l])
```

```python
l=[9,10,11.2]
```

```python
np.concatenate([a,b,l])
```

```python
l=[9,10,"11.2"]
```

```python
np.concatenate([a,b,l])
```

```python
l=(9,10) # tuple da verebiliriz
```

```python
np.concatenate([a,b,l])
```
