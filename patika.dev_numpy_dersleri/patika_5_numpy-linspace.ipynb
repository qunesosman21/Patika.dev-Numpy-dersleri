```python
import numpy as np
```

```python
# linspace farklı aralıklara böler

np.linspace(1,2) # default olarak 1 ve 2 arasında 50 noktaya böler.
```

# Kaç parçaya böleceğimizi belirlemek

```python
np.linspace(1,2,3) # 1 ve 2 arasında eşit olarak 3 parçaya böler. biz belirlemesek 50 parçaya böler
```

```python
np.linspace(1,2, num=3, endpoint=False) # 2 dahil etmeden 3 parçaya böler
```

```python
np.linspace(1.5, 4.7, 3) # float aralıklarda da böler. her zaman eşit olarak böler
```

```python
np.linspace(20,100,5, dtype=int)
```
