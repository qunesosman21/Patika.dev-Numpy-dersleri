```python
import numpy as np
```

# belirtilen bir elemandan array oluşturmak

```python
np.full((4,5), 5) #sadec 5 lerden oluşan bir array oluşturur
```

```python
np.full(4,5,5) # hata verir.
```

```python
np.full(4,5) # 4 elemanlı 5 lerden oluşan array
```

```python
np.full((4,5), 5.2, dtype= str) # integer a çevirip strin olarak yazar
```

```python
np.full((4,5), "5.2")
```

# Ardaşık sayılardan array oluşturmak

```python
np.arange(0,12) # 0 dan 12 kadar olan sayılardan oluşan bir array oluşturur. 12 dahi depğil
```

```python
np.arange(2,12)
```

```python
np.arange(2,12,2) # 2 den 12' e 2 artarak gider.
```

```python
np.arange(2,12,-2) # 2 den 12,e -2 olarak gidemediği için boş araray oluşturur.
```

```python
np.arange(12,2,-2) #12'den 2'e -2 azalarak gider
```
