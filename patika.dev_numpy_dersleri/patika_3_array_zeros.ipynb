```python
import numpy as np
```

# Elemanları 0'lardan oluşan array oluşturmak

```python
np.zeros(10) #default olarak bir şey belirlemezsek float olarak alır. 
```

```python
np.zeros(10, dtype="str") # stringin zerosu olarak boş string oluşturur
```

# Elemanlar 0'lardan oluşan multidimensional array oluşturmak


```python
# 3 row(satır) ve 4 colum(sütun)'u olan bir array oluşturmak

np.zeros((3,4))
```

```python
np.zeros(3,4)
```

```python
np.zeros([3,4])
```

```python
np.zeros([3,8,8])
```

```python
np.zeros([3,8,8], dtype="str")
```

```python
np.zeros([4,5,4], dtype="int")
```

```python
np.zeros(2, dtype= int)
```

```python
np.zeros((3,2), dtype=int)
```

```python
np.zeros((3,2), dtype= str) # "" ouluşan string array oluşturur
```

```python
np.zeros((2,3), bool)
```

```python
np.zeros((4,4), dtype="bool")
```

```python
np.zeros((4,4), bool)
```
